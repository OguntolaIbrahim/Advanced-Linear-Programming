```python
import numpy as np
import gurobipy as GRB
import gurobipy as gp
from gurobipy import GRB,quicksum
import time
```

# Q1. Stock Cutting Problem using Column Generation 

```python
#FINAL
tic=time.perf_counter()
n=8
size=100
stocking=np.zeros((8,8))
#length in inches
w=[7,11,15,19,23,27,35,42]
#Number ordered
b=[52,97,140,191,85,129,26,20]
df=[]

for i in range(n):
    stocking[i][i]=1
H=len(stocking)
#lenght of raws
fin_soln=100
eps=0.000000001


#stocking=np.vstack([stocking,[14,0,0,0,0,0,0,0]])
iterations=0

while(fin_soln>=1+eps):
    iterations+=1
    print('No of Stockings',len(stocking))
    
    #MASTER PROBLEM
    m = gp.Model()
    alpha=m.addVars(len(stocking),vtype=GRB.CONTINUOUS)
    m.update() # Update the model

    # Define the objective and constraints
    m.setObjective(quicksum(alpha[i] for i in range(len(stocking))) , GRB.MINIMIZE) # Define the objective function
    for i in range(len(b)):
        m.addConstr(quicksum(alpha[stock]*stocking[stock][i] for stock in range(len(stocking)) ) ==b[i])
    m.optimize() # Optimize
     # Extract the solution


    #Obtaining pi
    pi=m.Pi
    pi


    #SUBPROBLEM
    s=gp.Model()
    a=s.addVars(n,vtype=GRB.INTEGER)
    s.setObjective(quicksum(a[i]*pi[i] for i in range(len(w))),GRB.MAXIMIZE)
    s.addConstr(quicksum(w[i]*a[i] for i in range(n)) <= size )
    s.Params.MIPGap=0
    s.optimize()
    s.printAttr('X')
    fin_soln=s.objVal

    allvars =s.getVars()
    df=[]
    for v in allvars:
        df.append( v.X)
    print(fin_soln)
    print(m.objVal)
    stocking=np.vstack([stocking,df])

print(m.objVal)
print(m.printAttr('X'))
print(stocking)
iterations
toc=time.perf_counter()

```

```python
#Print Solution
stocking1=[]
print('No of Iterations : ', iterations)
for i in range(len(m.X)):
    if m.X[i]!=0:
        stocking1.append((stocking[i],m.X[i]))
for i,j in stocking1:
    print('stocking', i , 'alpha =',j) 
print('Lower Bound = ', m.objVal)
print('Runtime ' ,toc-tic, 'secs')
```
